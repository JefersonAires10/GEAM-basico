# Pymove: Bibiloteca Que Estuda Trajetórias

O Pymove é biblioteca de código aberto em Python, que surgiu em meados de 2018, oferece uma ampla gama de funcionalidades para análise e visualização de dados de trajétoria. Antes de explorar essas funcionalidades, devemos entender o conceito de trajetória e como esses dados podem ser trabalhados e as etapas realizadas pela pela biblioteca.

\\
**Trajétorias**

Uma trajetória pode ser definida como uma sequência de pontos cronologicamente ordenados, que representam o movimento de um objeto (pessoas, veículos e animais) em um espaço geográfico. Esses pontos são formados depa latitude longentude, data e horário. Os dados de trajetória são valiosos para entender o comportamento dos objetos em movimento e possibilitam insights importantes em diversos domínios.

\\
**Etapas**

A estrutura do Pymove é influenciada no paradgima criado por Yu Zheng, usada guia para realizar trabalhos de mineração de dados de trajetória. Temos os seguintes passos:

 1. Pré-processamento
 2. Indexação e recuperação de trajetórias
 3. Modelos
 4. Incerteza em trajetórias
 4. Vizualização de dados


Podemos ver mais de forma mais clara na figura abaixo:

Diagrama.JPG

Vamos explorar brevemente cada etapa:

# 1. Pré-processamento #
Nessa etapa são realizadas atividades com o objetivo de melhorar a qualidade do dados obter melhores resultados em processamentos e análises futuras. As técnicas mais conhecidas são:

-  Filtragem de ruído \\
-  Detecção de pontos de parada \\
-  Compressão  \\
-  Segmentação de trajetória \\
-  Map-matching \\


# 2. Indexação e recuperação de trajetórias #
Etapa é responsável pela adoção de técnicas eficazes para o gerenciamento desses dados, oferecendo a recuperação rápida dos dados. Existem dois principais tipos de consulta:

- KNN: recupera as K, primeiras trajetórias com a distância agregada mínima, para alguns pontos ou uma trajetória específica.
- Intervalo de consultas: recuperam os dados de trajetórias que estão contidos em um espaço ou intervalo.

# 3. Modelos #
Nessa etapa são aplicadas técnicas para extrair insights valiosos dos dados de trajetória, possibilitando o desenvolvimento de aplicações eficientes que fornecem informações instantâneas e relevantes. São exemplos dessas técnicas: classificação, detecção de anomalias e mineração de padrões.

# 4. Incerteza em trajetórias #
A localização de um indivíduo e/ou objeto podem ser incertos, ou até mesmo desconhecidos, entre dois pontos que descrevem sua trajetória., dois principais subcampos são abordados:

 1. Redução de incertezas: foca no desenvolvimento de técnicas para diminuir a incerteza presente entre dois pontos em uma trajetória. O objetivo é melhorar a precisão e a confiabilidade dos dados de localização,
 2. Preservação de privacidade: desenvolve técnicas que aumentam a incerteza das trajetórias sem comprometer a qualidade do serviço ou o uso dos dados, protegendo a privacidade dos usuários, dificultando a identificação precisa de suas trajetórias e evitando vazamentos de informações sensíveis.

# 5. Vizualização de Dados #
Responsável por facilitar a compreensão do comportamento de objetos em movimento. Essas visualizações podem ser combinadas com procedimentos de processamento dos dados, visando diminuir, limpar e filtrar os dados.

Para o primeiros teste, vamos usar "dict_data" com nossos MoveDataFrame, Um MoveDataFrame deve conter as colunas:

- lat: representa a latitude do ponto.
- lon: representa a longitude do ponto.
- datetime: representa a data e a hora do ponto.


Ao execultar a celula abaixo, iremos vizualizar as linhas do nosso data frame.

````python
import pymove as pm
from pymove import MoveDataFrame

list_data = [
    [39.984094, 116.319236, '2008-10-23 05:53:05', 1],
    [39.984198, 116.319322, '2008-10-23 05:53:06', 1],
    [39.984224, 116.319402, '2008-10-23 05:53:11', 1],
    [39.984224, 116.319402, '2008-10-23 05:53:11', 1],
    [39.984224, 116.319402, '2008-10-23 05:53:11', 1],
    [39.984224, 116.319402, '2008-10-23 05:53:11', 1]
]
move_df = MoveDataFrame(data=list_data, latitude="lat", longitude="lon", datetime="datetime", traj_id="id")
move_df.head()

````

Para trabalhar com mais dados, vamos usar uma função da biblioteca que permite abrir arquivos.

E para ver todas as informações desse arquuivo, vamos usar a função $\textrm{show_trajectories_info}()$

Vamos testar!

````python
df_move = pm.read_csv('geolife_sample.csv')


df_move.show_trajectories_info()
df_move.head()
````

Outra função fundamental é o $filter$ que permite filtrar dados, a célula a baixo contém um exemplo:

````python
from pymove import filters

bbox = (22.147577, 113.54884299999999, 41.132062, 121.156224)
filt_df = filters.by_bbox(df_move, bbox)
filt_df.head()
````
