<p>A Ciência de Dados possui três áreas principais:</p>

**<h3>1. Tradicional Aprendizado de Máquinas</h3>**


  ><p>Aprendizado de máquina é um subcampo da inteligência artificial que lida com algoritmos de computação que podem ser melhorados via dados de treinamento sem programação explícita. A união entre dados e algoritmos permite a inteligência artificial ler e interpretar padrões, além de aprender de forma autônoma, sem ação humana. Isso é possível graças ao machine learning, ou aprendizado de máquina.</p>
  
  >**<h3>Relação entre as áreas</h3>**
  A ciência de dados e a inteligência artificial são áreas inter-relacionadas que se complementam. Elas compartilham conceitos e ferramentas, especialmente no campo do aprendizado de máquina. Os cientistas de dados utilizam recursos de processamento e padronização da inteligência artificial em seu trabalho, ao mesmo tempo em que fornecem informações para o desenvolvimento do aprendizado de máquina. Embora as descrições de vagas para cientistas de dados e analistas de IA possam diferir, ambos os profissionais provavelmente utilizarão ferramentas de ciência de dados, como R e Python, em seu dia a dia.


  **<h3>2.  Internet das Coisas (IoT).</h3>**
  >**<h4>O que é IoT?</h4>**
  <p>IoT refere-se a dispositivos que não são computadores padrão, mas ainda têm a capacidade de transmitir dados. Isso inclui relógios inteligentes, sistemas de segurança residencial conectados à Internet, sistemas eletrônicos de cobrança de pedágio, construção de sistemas de gerenciamento de energia e muito, muito mais. Os dados da IoT são um ótimo recurso para projetos de ciência de dados!</p>

  >**<h4>O que é Big Data?</h4>**
  <p>Big Data é um termo utilizado para descrever o grande volume de dados (estruturados\não estruturados) gerados a cada segundo pelos mais diversos dispositivos. A grande vantagem do Big Data está atrelada à possibilidade de cruzamento desses dados por meio de diversas fontes, para, assim, angariar insights rápidos e de grande valor. Essa tecnologia possui várias aplicações em diferentes fatores, entre eles:
  1.   Marketing
  2.   Mercado Financeiro
  3.   Saúde


**<h4>Ligação entre as tecnologias</h4>**
><p>A coleta de dados por si só não gera vantagem competitiva. É necessário organizá-los em informações úteis com data analytics e ciência de dados. A inteligência artificial (IA) une IoT e Big Data, permitindo aprender com os dados e tomar decisões. Um exemplo é o uso de sensores geológicos para prevenir desastres. Com análise refinada e curadoria, obtém-se conhecimentos valiosos para decisões rápidas e eficazes. A IA possibilita respostas preditivas, evitando catástrofes e preservando recursos.</p>
  
**<h3>3.  Deep Learning (Aprendizado Profundo).</h3>**


  ><p>Aprendizagem profunda é um ramo do aprendizado de máquina que usa redes neurais artificiais para se aproximar da inteligência humana. Inspirada pelos neurônios humanos, a aprendizagem profunda usa a teoria dos grafos para organizar algoritmos ponderados em camadas de vértices e arestas. Os algoritmos da aprendizagem profunda são ótimos em processar dados não estruturados, como imagens ou linguagem.</p>

  **<h3>Redes neurais sem realimentação ou redes neurais feedforward (FF)</h3>**
  ><p>Os dados trafegam em uma direção desde a camada de entrada, passando pelas camadas ocultas, até a camada de saída; todos os vértices estão conectados e os dados nunca retomam o ciclo nas camadas ocultas. A rede feedforward é usada na compactação de dados e no processamento básico de imagens. </p>

  **<h3>Redes neurais recorrentes (RNN, Recurrent Neural Networks)</h3>**
  ><p>Um tipo de rede feedforward que adiciona um tempo de espera às camadas ocultas, o que permite o acesso à informações anteriores durante uma iteração corrente. Esse loop de feedback se aproxima da memória, o que torna as RNNs adequadas para processamento de linguagem.  Um bom exemplo disso é o texto preditivo que se baseia em palavras usadas com mais frequência para personalizar as sugestões.</p>

 **<h3>Redes neurais convolucionais (CCN, Convolutional neural networks)</h3>**
 ><p>Convolução é uma operação matemática em duas funções que produz uma terceira, descrevendo como uma é modificada pela outra. Usada principalmente para reconhecimento e classificação de imagens, as CNNs são os “olhos” da inteligência artificial. As camadas ocultas em uma CNN agem como filtros matemáticos usando a soma ponderada para identificar arestas, cores, contraste e outros elementos de um pixel.</p>
